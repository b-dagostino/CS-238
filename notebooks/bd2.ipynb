```python
def update(b, o, T, O):
    """
    Perform a belief update given prior belief b, observation o,
    transition matrix T, and observation emission matrix O.
    """
    b_prime = np.zeros_like(T_final[0])
    for i, s in enumerate(S.index):
        po = O[s, o]
        sum_term = 0
        for j, s_prev in enumerate(S.index):
            sum_term += T[s_prev, s] * b[j]
        b_prime[i] = po * sum_term

    # Normalize
    b_upd = b_prime / np.sum(b_prime)
        
    return b_upd
```

```python
b_0 = np.zeros(len(S))
# Uniform over non-event states
num_non_event_states = len(S[S.e == "⊥"])
for s in S[S.e == "⊥"].index:
    b_0[s] = 1 / num_non_event_states
b_0
b = b_0
i = 0
```

```python
b = update(b, 0, T_final, O_0).round(3)
i += 1
print(f"After observation {i}:")
display(b.round(3))
argmax_state = np.argmax(b)
display(S.iloc[argmax_state].to_frame().T)
```
